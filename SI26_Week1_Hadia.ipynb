Q1. What is OCR?

OCR (Optical Character Recognition) is a technology that converts text from images into editable digital text. It helps computers read printed or handwritten text. OCR is widely used for document digitization and automation.

Q2. Why is Urdu OCR harder?

Urdu OCR is more difficult because Urdu is written from right to left and many letters change their shape depending on their position. Characters are connected, making recognition challenging. Different fonts and handwriting styles also increase the difficulty.

Q3. Real World Uses

Urdu OCR can be used to digitize old Urdu books and newspapers. It is also useful for converting printed government documents into searchable digital files.

```python
!pip install Pillow arabic-reshaper python-bidi

from PIL import Image, ImageDraw
import arabic_reshaper
from bidi.algorithm import get_display
import os

urdu_texts = [
    "پاکستان زندہ باد",
    "آج کا موسم خوشگوار ہے",
    "تعلیم ہر انسان کا حق ہے",
    "کراچی پاکستان کا سب سے بڑا شہر ہے",
    "محنت کامیابی کی کنجی ہے"
]

os.makedirs("data/synthetic", exist_ok=True)

for i, text in enumerate(urdu_texts):
    reshaped = arabic_reshaper.reshape(text)
    bidi_text = get_display(reshaped)

    img = Image.new("RGB", (600, 100), color="white")
    draw = ImageDraw.Draw(img)

    draw.text((20, 30), bidi_text, fill="black")

    img.save(f"data/synthetic/urdu_{i+1}.png")

print("Done!")
```
